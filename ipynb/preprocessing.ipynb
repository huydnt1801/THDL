```python
all_data_folder = prj_path + "/crawl/crawl/spiders/all_data"
```

```python
muaban_csv = pd.read_csv(all_data_folder + "/muaban.csv").drop(columns=['Unnamed: 0'], axis=1)
okxe_csv = pd.read_csv(all_data_folder + "/okxe.csv").drop(columns=['Unnamed: 0'], axis=1)
xemayhoangkien_csv = pd.read_csv(all_data_folder + "/xemayhoangkien.csv").drop(columns=['Unnamed: 0'], axis=1)
webike_csv = pd.read_csv(all_data_folder + "/webike.csv").drop(columns=['Unnamed: 0'], axis=1)
```

```python
def convert_price(x): 
    try :
        return int(re.sub('\s+', '', x)[0:-1].replace('.', '').strip())
    except:
        return ''
```

```python
webike_csv['price'] = webike_csv['price'].fillna(0)
webike_csv['address'] = webike_csv['Địa điểm'].apply(lambda x: re.sub('\s+', ' ', x) if type(x)==str else x)
webike_csv = webike_csv.fillna('-')
```

```python
# muaban_csv['Số km đã đi'] = muaban_csv['Số km đã đi'].apply(lambda x: f'0-{x}km' if type(x) == str else 0)
# muaban_csv = muaban_csv.drop(columns=['Giá'])
muaban_csv = muaban_csv.fillna('-')
```
