```python
import pandas as pd
import json
```

```python
data_folder = "./crawl/crawl/spiders/data"
```

```python
muaban_csv = pd.read_excel(data_folder + "/muaban.xlsx")
okxe_csv = pd.read_excel(data_folder + "/okxe.xlsx")
webike_csv = pd.read_excel(data_folder + "/webike.xlsx")
xemayhoangkien_csv = pd.read_excel(data_folder + "/xemayhoangkien.xlsx")
```

```python
dataframes = [
    muaban_csv,
    okxe_csv,
    webike_csv,
    xemayhoangkien_csv
]
```

```python
info = json.loads(muaban_csv['information'][0])
```

```python
for i in range (10):
    info = json.loads(muaban_csv['information'][i])
    print("--------------")
    for key in info.keys():
        print(info[key])
```

```python
for df in dataframes:
    columns = set(df.columns)
    for i in range(len(df)):
        info_json = json.loads(df['information'][i])
        
        for key in info_json.keys():
            if key in columns:
                df[key][i] = info_json[key]
            else: 
                df[key] = [''] * len(df)
                df[key][i] = info_json[key]
```
